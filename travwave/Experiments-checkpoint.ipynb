```python
from __future__ import division
from navigation import *
from solver import *
from equation import *
from boundary import *

size = 256
length = np.pi
equation = Whitham(size, length)
boundary = ConstZero()
solver = Solver(equation, boundary)
nav = Navigator(solver.solve)
initial_guess = equation.compute_initial_guess()
initial_velocity = equation.bifurcation_velocity()
p1 = (initial_velocity, 0)
epsilon = .01
p0 = (initial_velocity, -epsilon)
nav.initialize(initial_guess, p1, p0)
```

```python
lambda1 = np.log(5)
ampl = 0.2
scale = equation.length/np.pi
argum = np.log((1 + np.sqrt(0.5))/np.sqrt(0.5))
flambda = 2/lambda1*argum
eta = ampl*1/np.cosh(flambda*scale*equation.nodes)**2
C = 1/(equation.length)*ampl/flambda*2*np.tanh(equation.length/flambda)
eta = eta - C
```

```python
nav.run(75)
print('Done')
```
