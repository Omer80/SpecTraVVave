```python
from __future__ import division
from navigation import *
from solver import *
from equation import *

size = 64
length = np.pi
equation = KDV(size, length)
solver = Solver(equation)
nav = Navigator(solver.solve)
initial_guess = equation.compute_initial_guess()
initial_velocity = equation.bifurcation_velocity()
p1 = (initial_velocity, 0)
epsilon = .01
p0 = (initial_velocity, -epsilon)
nav.initialize(initial_guess, p1, p0)
```

```python
len(initial_guess)
```

```python
nav.run(30)
print('Done')
```

```python
import dynamic_code
xx = np.arange(0, 2*length, length/size)
u = nav.store[-1][0]
velocity = nav.store[-1][1][0]
dyn = dynamic_code.Dynamic_code(equation, u, velocity)
uu = dyn.interpolation()
t_wave = dyn.evolution(solution = uu, dt = 0.001, periods = 2)
```

```python
plot(t_wave)
plot(uu)
```
